```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn import linear_model
```

# Creating dataframe

```python
data = {
    'age' : [25,30,35,40,45],
    'premium': [18000, 32000, 40000, 47000, 55000]
}
df = pd.DataFrame(data)
df
```

```python
sns.lmplot(x='age', y='premium', data=df)
```

```python
model = linear_model.LinearRegression()
```

```python
df['age']
```

```python
y = df['premium']
y
```

```python
model.fit(df[['age']], df[['premium']])  #takes data as double array
```

```python
model.predict([[32]])
```

```python
model.predict([[41]])
```

```python
model.coef_  #value of slope
```
