```python
from sklearn import datasets
```

```python
iris_data = datasets.load_iris()
```

```python
iris_data
```

```python
iris_data['data']
```

```python
iris_data['target']
```

```python
iris_data.target_names
```

```python
iris_data.feature_names
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'sepal length': iris_data.data[:, 0],
    '': iris_data.data[:, 1],
    'petal length': iris_data.data[:, 2],
    'petal width' : iris_data.data[:, 3],
    'target' : iris_data.target
})
```

```python
df
```

```python
x = df[['sepal length', 'sepal width', 'petal length', 'petal width']]  # features
y = df[['target']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_test, x_train, y_test, y_train = train_test_split(x,y, train_size=0.2)
```

```python
x_train
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators=100, criterion='gini')
```

```python
classifier.fit(x_train, y_train)
```

```python
classifier.predict(x_test)
```

```python
y_test
```

```python
classifier.score(x_test, y_test)
```

```python
imp_features = pd.Series(classifier.feature_importances_, index=iris_data.feature_names).sort_values(ascending=False)
```

```python
imp_features
```
